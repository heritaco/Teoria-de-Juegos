# Core facts

* **Normal-form game:** players (N), pure strategies ({S_i}), payoff functions (v_i:S_1\times\cdots\times S_n\to\mathbb{R}). Simultaneous choice means choosing without observing others. Two-player finite games admit matrix form.
* **Strategy, profile:** pure strategy is a deterministic plan; profile (s=(s_1,\dots,s_n)). 
* **Complete information + common knowledge** underpin static strategic analysis.
* **Dominance:** never play a strictly dominated strategy; strict dominant-strategy equilibrium (if it exists) is unique. Weak dominance need not be unique.
* **Nash equilibrium (pure):** each (s_i^*) is a best response to (s_{-i}^*). Practical 3-step “underline/overline” matrix procedure. A strict-dominant profile or the unique IESDS survivor is the unique NE.
* **Pareto:** (s) Pareto-dominates (s') if all are weakly better and someone strictly better. NE need not be Pareto-optimal.

# Solve-any-2×2 checklist

1. Delete strictly dominated rows/columns. Stop if one remains. 
2. Best-response marks: underline row-player maxima in each column; overline column-player maxima in each row. Intersections are pure NE. 
3. Compare NE by Pareto dominance if asked about efficiency. 
4. Zero-sum special: check for a **saddle point** (row minimum maximized and column maximum minimized). That cell is the value. 

# Canonical examples you likely saw

* **Prisoner’s Dilemma:**
  [
  \begin{array}{c|cc}
  & C & D\\hline
  C &(-1,-1)&(-10,0)\
  D &(0,-10)&(-9,-9)
  \end{array}
  ]
  Strictly dominant (D) for both (\Rightarrow) NE ((D,D)). Inefficient vs ((C,C)).
* **Battle of the Sexes:** two pure NE ((A,A)) and ((B,B)); no dominance.
* **Matching Pennies (zero-sum):** no pure NE; use mixed (not in your slides’ excerpt, but treat with minmax idea from saddle-point discussion). 
* **Bismarck Sea (zero-sum):** matrix (\begin{small}\begin{pmatrix}2&2\1&3\end{pmatrix}\end{small}) has saddle at ((\text{North},\text{North})) with value (2).
* **Stag Hunt (Rousseau):** two pure NE ((S,S)) and ((H,H)). ((S,S)) Pareto-dominates. Coordination vs risk. 
* **Cournot duopoly (no costs, (p=1-Q)):** best responses yield (q_1=q_2=\tfrac13) (Cournot NE), not Pareto-optimal; each at (\tfrac14) would be better.
* **Tragedy of the Commons:** (v_i=\ln k_i+\ln!\big(K-\sum_j k_j\big)). Best responses give (k_i=\tfrac{K-\sum_{j\neq i}k_j}{2}). For (n=2): (k_1=k_2=K/3) (NE) vs Pareto-optimum (k_1=k_2=K/4).

# Rapid drills (with answers)

1. **Find NE:**
   (\begin{small}\begin{array}{c|cc} & L & R\\hline U&(3,3)&(0,2)\ D&(2,0)&(1,1)\end{array}\end{small})
   **Ans:** ((U,L)).
2. **Delete dominated strategy:**
   (\begin{small}\begin{array}{c|cc} & L & R\\hline U&(1,4)&(1,1)\ D&(0,3)&(0,0)\end{array}\end{small})
   **Ans:** (D) strictly dominated by (U); NE ((U,L)).
3. **Zero-sum saddle?**
   (\begin{small}\begin{array}{c|cc} & L & R\\hline U&2&4\ D&3&1\end{array}\end{small})
   **Ans:** Row minima: (\min(U)=2,\min(D)=1\Rightarrow\max=2). Column maxima: (\max(L)=3,\max(R)=4\Rightarrow\min=3). No saddle (2≠3).
4. **BoS-type:**
   (\begin{small}\begin{array}{c|cc} & A & B\\hline A&(2,1)&(0,0)\ B&(0,0)&(1,2)\end{array}\end{small})
   **Ans:** ((A,A)) and ((B,B)).
5. **PD-type:**
   (\begin{small}\begin{array}{c|cc} & C & D\\hline C&(4,4)&(0,5)\ D&(5,0)&(1,1)\end{array}\end{small})
   **Ans:** (D) strictly dominant; NE ((D,D)); ((C,C)) Pareto-dominates.

& L & R\\hline
U&2&4\

&\max\min \text{ (row)}=\max{2,1}=2,\
